```python
import matplotlib.pyplot as plt
import networkx as nx
from nxviz.plots import CircosPlot
```

```python
import random
```

```python
G = nx.karate_club_graph()
```

```python
nx.set_node_attributes(G, dict(zip(G.nodes, [random.randint(1, 4) for n in G.nodes])), "tipo_socio")
```

```python
nx.set_node_attributes(G, dict(zip(G.nodes, [random.randint(1, 4) for n in G.nodes])), "size")
```

```python
c = CircosPlot(
    G,
    dpi=600,
#     node_grouping="weight",
#     edge_width="weight",
    figsize=(20, 20),
    node_color="tipo_socio",
    node_labels=True,
    node_size="size"
)
c.draw()
plt.show()
```
