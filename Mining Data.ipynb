## Identificação da grande categoria de ensino em que o CNPJ se enquadra

```python
FUNDAMENTAL_CODE = '851'
MEDIO_CODE = '852'
SUPERIOR_CODE = '853'
TECNICO_CODE = '854'
APOIO_CODE = '855'
OUTRAS_CODE = '859'

CLASSES = [FUNDAMENTAL_CODE, MEDIO_CODE, SUPERIOR_CODE, TECNICO_CODE, APOIO_CODE, OUTRAS_CODE]
```

```python
df_main = pd.read_csv(EDU_COMPANIES_FILE, usecols=COMPANY_COLUMNS)
df_main.head()
```

```python
df_second = pd.read_csv(SECOND_EDU_COMPANIES_FILE, usecols=COMPANY_COLUMNS)
df_second.head()
```

```python
df_companies = pd.concat([df_main, df_second]).drop_duplicates()
```

```python
df_companies
```

### Mapeamento do tipo de atividade segundo os CNAEs secundários¶

```python
df_cnaes = pd.read_csv(EDU_CNAES_FILE, index_col=0)
df_cnaes.head()
```

```python
for cls in CLASSES:
    if not cls in df_cnaes.columns:
        df_cnaes[cls] = 0
    df_cnaes[cls] = df_cnaes[cls] + df_cnaes.cnae.apply(lambda x: 1 if str(x).startswith(cls) else 0)
```

```python
df_cnaes.head()
```

```python
df_grouped_cnaes = df_cnaes.drop(columns=['cnae_ordem', 'cnae']).groupby('cnpj', as_index=False).sum()
```

```python
df_grouped_cnaes
```

```python
df_only_main = df_companies[~df_companies.cnpj.isin(df_grouped_cnaes.cnpj)][['cnpj']] 
```

```python
for cls in CLASSES:
    if not cls in df_only_main.columns:
        df_only_main[cls] = 0
```

```python
df_only_main
```

```python
df_all_grouped_cnaes = pd.concat([df_only_main, df_grouped_cnaes]).drop_duplicates().reset_index(drop=True)
```

```python
df_all_grouped_cnaes
```

```python
df_companies = df_companies.merge(df_all_grouped_cnaes, how='left', on='cnpj')
```

```python
df_companies
```

### Mapeamento do tipo de atividade segundo o CNAE principal
